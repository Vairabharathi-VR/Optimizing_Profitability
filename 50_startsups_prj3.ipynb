```python
from scipy import stats
f_statistic, p_value = stats.f_oneway(df['Profit'][df['State'] == 'New York'],
                                      df['Profit'][df['State'] == 'California'],
                                      df['Profit'][df['State'] == 'Florida'])

print(f'F-statistic: {f_statistic}')
print(f'P-value: {p_value}')
```

```python
# Based on the one-way ANOVA test, there is no evidence to suggest that the categorical variable 
# "State" has a significant impact on the numerical variable "Profit" across the states of 
# New York, California, and Florida.
```

# Model planning

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score, mean_squared_error
from math import sqrt
lr = LinearRegression()
```

```python
x = df[['R&D Spend', 'Marketing Spend']]
x.head()
```

```python
y = df['Profit']
y.head()
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y, train_size=0.8)
```

```python
x_train.shape, y_train.shape, x_test.shape, y_test.shape
```

```python
model_train = lr.fit(x_train, y_train)
```

```python
lr.coef_
```

```python
lr.intercept_
```

```python
pred = model_train.predict(x_test)
```

```python
df2 = pd.DataFrame(pred, columns=['Predicted'])
```

```python
df2['Actual'] = y_test.values
```

```python
df2
```

```python
r2 = r2_score(df2['Actual'], df2['Predicted'])
print("The Strength of the model is - ", round(r2,2))
```

```python
print(f'The Accuracy of the model is {round(r2*100,2)} %, so the prediction with this model is good')
```
